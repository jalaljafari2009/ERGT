This notebook runs real training, not synthetic telemetry. It is the first Colab-facing guarded run after the ERGT-03 wrapper smoke.

Execution shape:

```text
baseline: full profile

real/random/shuffled/no-memory/instantaneous: full adaptive-alpha training
```

Default output bundle: `ergt_04_guarded_adaptive_training_report_bundle.zip`

Default local review path after download: `C:\Users\Administrator\Downloads\ergt_04_guarded_adaptive_training_report_bundle.zip`

The bundle excludes checkpoints and model weights. Live training output is streamed during each condition with `python -u`.


````python
from __future__ import annotations

import os
import subprocess
import sys
from pathlib import Path

GIT_REPO_URL = "https://github.com/jalaljafari2009/ERGT.git"
COLAB_PROJECT_ROOT = Path("/content/ERGT")


def prepare_project_root() -> Path:
    current_root = Path.cwd()
    if (current_root / "experiments").exists() and (current_root / "models").exists():
        return current_root
    if Path("/content").exists():
        if not COLAB_PROJECT_ROOT.exists():
            subprocess.run(["git", "clone", "--depth", "1", GIT_REPO_URL, str(COLAB_PROJECT_ROOT)], check=True)
        elif (COLAB_PROJECT_ROOT / ".git").exists():
            subprocess.run(["git", "-C", str(COLAB_PROJECT_ROOT), "pull", "--ff-only"], check=False)
        if (COLAB_PROJECT_ROOT / "experiments").exists():
            os.chdir(COLAB_PROJECT_ROOT)
            return COLAB_PROJECT_ROOT
    raise RuntimeError(f"ERGT project root was not found. Expected {GIT_REPO_URL}.")


PROJECT_ROOT = prepare_project_root()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(f"Project root: {PROJECT_ROOT}")
````
